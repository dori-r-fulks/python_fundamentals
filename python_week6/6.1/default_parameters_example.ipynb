```python
#Function with default parameter
def greet(name, greeting="Hello"):
    print(f"{greeting}, {name}!")
```

```python
#Call different ways
#Hello, Alice! (uses default)
greet("Alice")
```

```python
#Hi, Bob! (overrides default)
greet("Bob", "Hi")
```

```python
#Good morning, Carol! (overrides default)
greet("Carol", "Good morning")
```

```python
greet("Dallas")
```

```python
greet("Erik", "g'Day!")
```
