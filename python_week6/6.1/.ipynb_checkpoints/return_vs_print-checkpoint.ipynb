```python
#Function that PRINTS
def add_print(a,b):
    result = a + b
    print(result)
```

```python
#Function that RETURNS
def add_return(a, b):
    result = a + b
    return result
```

```python
#Try to use them:
x = add_print(5, 3)
```

```python
print(x)
```

```python
y = add_return(5, 3)
```

```python
print(y)
```
