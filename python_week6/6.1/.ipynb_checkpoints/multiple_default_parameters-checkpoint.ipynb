```python
#Multiple defaults
def create_user(name, age=18, city="unknown"):
    print(f"Name: {name}")
    print(f"Age: {age}")
    print(f"City: {city}")
```

```python
#Different ways to call:
create_user("Alice")
```

```python
create_user("Bob", 25)
```

```python
create_user("Carol", 30, "Boston")
```

```python
create_user("Dallas", 27, "Los Angeles")
```

```python
create_user("Bob", 27, "NYC")
```
