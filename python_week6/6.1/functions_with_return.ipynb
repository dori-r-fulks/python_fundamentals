```python
#Function that returns value
def add(a, b):
    result = a + b
    return result
```

```python
#Call and store the result
sum1 = add(5, 3)
print(sum1)
```

```python
sum2 = add(10, 20)
print(sum2)
```

```python
#use returned value in calculation
total = (add(593, 3) + add(10, 29)) * 2
print(total)
```

```python
total1 = sum1 + sum2
print(total1)
```

```python
total2 = total + total1
print(total2)
```
