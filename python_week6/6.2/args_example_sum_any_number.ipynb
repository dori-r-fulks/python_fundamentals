```python
#Functions accepts any number of arguments
def add_all(*args):
    print(f"Received: {args}") #Tuple
    return (sum(args) % (2)) 
```

```python
#Call with different amounts
print(add_all(1,2))
```

```python
print(add_all(1,2,3,4,5))
```

```python
print(add_all(10,20,30,40,50,60))
```
