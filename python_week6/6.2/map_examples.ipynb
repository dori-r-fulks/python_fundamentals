```python
numbers = [1, 2, 3, 4, 5]
```

```python
#Square each number
squared = list(map(lambda x: x ** 2, numbers))
print(squared)
```

```python
#convert to strings
strings = list(map(str, numbers))
print(strings)
```

```python
#Multiple lists (zip them)
a = [1, 2, 3]
b = [10, 20, 30]
sums = list(map(lambda x, y: x + y, a, b))
print(sums)
```
