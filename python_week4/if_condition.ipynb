```python
#Check if passing grade
grade = 120
if grade >= 60 and grade <= 100:
    print("You passed!")
    print("Congratulations!")
if grade < 60:
    print("You did not pass.")
    print("Good luck on your retake!")
if grade > 100:
    print("Your score is invalid, please see administrator for assistance.")
    
print("Grade check complete.")
```
