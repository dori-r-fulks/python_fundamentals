```python
#Categoeize people by age
ages = [5, 25, 15, 67, 45, 12, 70, 30, 8, 55]
children = []
adults = []
seniors = []
for age in ages:
    if age < 18:
        children.append(age)
    elif age < 65:
        adults.append(age)
    else:
        seniors.append(age)
print(f"Children: {children}")
print(f"Adults: {adults}")
print(f"Seniors: {seniors}")
```
