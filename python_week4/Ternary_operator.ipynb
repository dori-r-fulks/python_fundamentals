```python
#1. Grade status in one line
score = 85
status = "PASS" if score >= 60 else "FAIL"
print(status)

#2. Age category
age = 25
category = "adult" if age >= 18 else "minor"
print(category)

#3. Even/odd
num = 70
parity = "even" if num %2 == 0 else "odd"
print(parity)

#4. Max of two numbers
a, b = 100, 15
max_value = a if a > b else b
print(max_value)
```
