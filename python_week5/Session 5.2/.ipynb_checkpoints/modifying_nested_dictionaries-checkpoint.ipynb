```python
students = {
    "Alice": {"age": 20, "gpa": 3.8}
}
```

```python
#Add new student (outer level)
students["Bob"] = {"age": 21, "gpa": 3.6}
```

```python
#Add field to existing student (inner level)
students["Alice"]["major"] = "CS"
```

```python
#Update nested value
students["Alice"]["gpa"] = 3.9
```

```python
print(students["Alice"])
```

```python
students["Alice"]["graduation year"] = 2029
students["Bob"]["graduation year"] = 2030
students["Bob"]["graduation year"] = "marketing"
```

```python
print(students["Bob"])
```
