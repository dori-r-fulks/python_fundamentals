```python
student = {"name": "Bob", "age": 20}
```

```python
#update with new data
new_data = {"major": "CS", "gpa": 3.8}
student.update(new_data)
print(student)
```

```python
new_data = {"school": "GIT", "city": "Atlanta"}
student.update(new_data)
print(student)
```

```python
#pop (remove and return)
age = student.pop("age")
print(f"Removed age: {age}")
print(student)
```

```python
#clear all
student.clear()
print(student)
```

```python
new_data = {"Favorite food": "Icecream", "Favorite color": "Green"}
student.update(new_data)
print(student)
```
