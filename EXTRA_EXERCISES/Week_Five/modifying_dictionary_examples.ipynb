```python
student = {"name": "Bob", "age": 20}
```

```python
#Add new field
student["major"] = "Computer Science"
print(student)
```

```python
#update existing field
student["age"] = 21
print(student["age"])
```

```python
#delete a field
del student["age"]
print(student)
```
