```python
#Calculate statistics from dictionary
prices = {
    "laptop": 999.99,
    "mouse": 25.50,
    "keyboard": 79.99,
    "monitor": 299.99
}
```

```python
#Total value
total = sum(prices.values())
print(f"Total: ${total:.2f}")
```

```python
#Average price
avg = total / len(prices)
print(f"Average: ${avg:.2f}")
```

```python
#Most expensive
max_item = max(prices, key=prices.get)
print(f"Most expensive: {max_item}")
```

```python
#Highest price
highest = max(prices.values())
print(f"Highest Price: {highest}")
```
