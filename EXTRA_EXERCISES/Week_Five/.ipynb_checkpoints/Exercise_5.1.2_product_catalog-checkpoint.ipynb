```python
product1 = {
    "name": "yellowtail",
    "price": 79.99,
    "category": "grocery",
    "in_stock": True
}
product2 = {
    "name": "gymshoes",
    "price": 156.00,
    "category": "clothes",
    "in_stock": True
}
product3 = {
    "name": "lotion",
    "price": 14.69,
    "category": "health and beauty",
    "in_stock": False
}
products = [product1, product2, product3]
total = sum(product["price"] for product in products)
in_stock_count = sum(products["in_stock"] for products in products)
highest = max(product["price"] for product in products)
for product in products:
    product["discounted_price"] = product["price"] * 0.90
    print(f"{product["name"]}\nOriginal price: {product["price"]:.2f} --> Discount Price: {product["discounted_price"]:.2f}\n")
print(f"Total products in stock: {in_stock_count}")
print(f"Highest priced item: {highest:.2f}")
```
