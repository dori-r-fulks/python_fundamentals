```python
contacts = {
    "Alice":{"phone": "555-1234", "email": "alice@email.com"},
    "Bob": {"phone": "555-5678", "email": "bob@email.com"}
}
```

```python
new_contact = {
    "Charles": {"phone": "555-2345", "email": "charles@email.com"},
    "Davis": {"phone": "555-3456", "email": "davis@email.com"}
}
contacts.update(new_contact)
print(contacts)
```

```python
new_contact = {
    "Elena": {"phone": "555-4567", "email": "elena@email.com"}
}
contacts.update(new_contact)
print(contacts)
```

```python
contacts["Elena"]["phone"] = "555-6789"
```

```python
print(contacts)
```

```python
if "email" in contacts["Elena"]:
    print(contacts["Elena"]["email"])
if "phone" in contacts["Elena"]:
    print(contacts["Elena"]["phone"])
if "address" in contacts["Elena"]:
    print(contacts["Elena"]["address"])
else:
    print("Info not provided")
```

```python
for name in contacts:
    print(name)
```

```python
total = len(contacts)
print(total)
```

```python
contact3 = contacts.pop("Charles")
print(f"Removed contact: {contact3}")
```
