```python
grades = {
    "math": 85,
    "english": 92,
    "science": 88,
    "history": 90
}
total = sum(grades.values())
avg = total / len(grades)
highest_subject = max(grades, key=grades.get)
lowest_subject = min(grades, key=grades.get)
subject = (grades.items())
highest = grades[highest_subject]
lowest = grades[lowest_subject]
print(f"Average: {avg:.2f}")
print(f"Highest Grade: {highest:.2f} Subject: {highest_subject}")
print(f"Lowest Grade: {lowest:.2f} Subject: {lowest_subject}")
```
