```python
#create dict of squares
squares = {x: x**2 for x in range(1,6)}
print(squares)
```

```python
#double all values
prices = {"apple": 1.50, "banana": 0.75, "kiwi": 3.89, "plantain": 1.25}
doubled = {k: v*2 for k, v in prices.items()}
print(doubled)
```

```python
#filter high grades only
grades = {
    "math": 85,
    "english": 92,
    "science": 78,
    "french": 99,
    "biology": 80
}
high = {k: v for k, v in grades.items() if v >= 80}
print(high)
```
