```python
#Method 1: read() - whole file
with open("sample.txt", "r") as file:
    content = file.read()
    print(type(content))
```

```python
#Method 2: readlines() - list of lines
with open("sample.txt", "r") as file:
    lines = file.readlines()
    print(type(lines))
    print(lines[2])
```

```python
#Method 3: Loop (Best for large files)
with open("sample.txt", "r") as file:
    for line in file:
        print(line.strip())
```
