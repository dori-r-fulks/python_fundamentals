```python
count = 0
with open("test.txt", "r") as file:
    for line in file:
        if line.strip():
            count += 1
print(count)
```

```python
word = "Comment"
count = 0
with open("test.txt", "r") as f:
    for line in f:
        if word in line:
            count += 1
print(count)
```

```python
filename = "test.txt"
word = "Comment"
total_lines = 0
non_empty_lines = 0
word_line = 0

with open (filename, "r") as f:
    for line in f:
        total_lines += 1
        if line.strip():
            non_empty_lines += 1
        if word in line:
            word_line += 1
print("-----File Statistics-----")
print(f"File: '{filename}'")
print(f"Total Lines: {total_lines}")
print(f"Non-empty Lines: {non_empty_lines}")
print(f"Lines containing '{word}': {word_line}")
```
