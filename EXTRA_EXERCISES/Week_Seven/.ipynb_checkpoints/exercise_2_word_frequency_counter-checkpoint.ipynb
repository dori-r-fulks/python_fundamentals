```python
with open("word_frequency.txt", "w") as f:
    f.write("Python is great Python is powerful Python is fun")
```

```python
with open("word_frequency.txt", "r") as file:
    content = file.read()
    print(content)
```

```python
with open("word_frequency.txt", "r") as f:
    text = f.read()
    words = text.split()
print(words)
```

```python
with open("word_frequency.txt", "r") as f:
    words =  text.split()
    frequency = {}
for word in words:
    frequency[word] = frequency.get(word, 0) + 1
print(frequency)
```

```python
top_5 = sorted(frequency.items(), key=lambda x: x[1], reverse=True)[:5]
print("Top 5 most common words:")
for word, count in top_5:
    print(f"{word}: {count}\n")
```
