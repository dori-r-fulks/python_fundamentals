```python
import csv
```

```python
students = [
    ["Name", "Age", "Score"],
    ["Alice", 25, 85],
    ["Bob", 30, 92],
    ["Carol", 28, 78]
]
```

```python
with open("students.csv", "w") as f:
    for row in students:
        f.write(",".join(map(str, row)) + "\n")
```

```python
total_score = 0
count = 0
highest_score = 0
top_student = ""

with open("students.csv", "r") as f:
    header = f.readline()
    
    for line in f:
        parts = line.strip().split(",")
        
        name = parts[0]
        age = int(parts[1])
        score = int(parts[2])
        
        total_score += score
        count += 1
        
        if score > highest_score:
            highest_score = score
            top_student = name

average_score = total_score / count
print("----- Student Statistics -----")
print(f"Average Score: {average_score:.2f}")
print(f"Highest Scorer: {top_student} ({highest_score})")
```
