```python
#Function with two parameters
def introduce(name, age):
    print(f"Hi, my name is {name}")
    print(f"I am {age} years old")
```

```python
#Call with two arguments
introduce("Aaliyah", 22)
```

```python
introduce("Breezy", 78)
```

```python
introduce("Car'lann", 17)
```

```python
def introduce(name, age, city, state, zip):
    print(f"Hi my name is {name}. I am {age} and I live in {city}, {state} {zip}")
```

```python
introduce("Panda", 17, "Chicago", "IL", 60606)
```

```python
def profile(name, age, major, favorite_food):
    print(f"{name} is {age} years old, majors in {major}, and loves to eat {favorite_food}.")
```

```python
profile("Dorifu", 29, "Data Science", "icecream")
```

```python
profile("Zach", 66, "Fire Technology", "cookies")
```
