```python
#Function that returns value
def add(a, b):
    result = a + b
    return result
```

```python
#Call and store result
sum1 = add(5, 3)
print(sum1)
```

```python
sum2 = add(10, 20)
print(sum2)
```

```python
total = add(5, 3) + add(10, 2)
print(total)
```

```python
total = sum1 + sum2
print(total)
```

```python
def multi(a, b):
    result = a * b
    return result
sum3 = multi(2, 3)
print(sum3)
```

```python
total = multi(2, 3) + multi(2, 3)
print(total)
```
