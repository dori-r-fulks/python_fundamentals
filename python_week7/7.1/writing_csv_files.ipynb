```python
import csv
```

```python
#data to save
students = [
    ["Name", "Age", "Score"],
    ["Alice", 25, 95],
    ["Bob", 30, 87],
    ["Carol", 28, 92]
]
```

```python
#Write CSV file
with open("students.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(students)
```
