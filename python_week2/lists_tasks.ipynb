```python
tasks = ['email', 'meeting', 'lunch']
```

```python
#Add to end
tasks.append('report')
print(tasks)
```

```python
#add at specific position
tasks.insert(1, 'coffee')
print(tasks)
```

```python
#Remove specific item
tasks.remove('lunch')
print(tasks)
```

```python
#remove last item
last = tasks.pop()
print(last)
print(tasks)
```
