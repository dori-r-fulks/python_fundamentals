```python
scores = [95, 87, 92, 78, 88, 91]
```

```python
count = len(scores)
print(f'Number of scores: {count}')
```

```python
total = sum(scores)
print(f'Total points: {total}')
```

```python
average = total / count
print(f'Average: {average:.1f}')
```

```python
print(f'Highest: {max(scores)}')
print(f'Lowest: {min(scores)}')
```
