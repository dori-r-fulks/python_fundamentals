```python
numbers = [42, 17, 93, 8, 55]
```

```python
#Sort in ascending order (modifies list)
numbers.sort()
print(numbers)
```

```python
#Reverse the order
numbers.reverse()
print(numbers)
```

```python
#Sort names alphabetically
names = ['Charlie', 'Alice', 'Bob']
names.sort()
print(names)
 
```
